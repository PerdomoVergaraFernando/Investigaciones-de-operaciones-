## Se desea analizar cómo evolucionan las probabilidades de estar en cuatro sectores $(x_n, y_n, z_n, t_n )$ diferentes  a lo largo del tiempo.
##Cada sector representa un estado dentro de un sistema dinámico donde las probabilidades de pasar de un estado a otro están definidas por una matriz de transición $A$.

###El modelo está dado por la ecuación matricial:
$$X_n = A \cdot X_{n-1}$$

##Donde:

$$A =
\begin{pmatrix}
\frac{1}{4} & 0 & \frac{1}{4} & \frac{2}{4} \\
\frac{2}{4} & \frac{1}{4} & 0 & \frac{1}{4} \\
\frac{1}{4} & \frac{2}{4} & \frac{1}{4} & 0 \\
0 & \frac{1}{4} & \frac{2}{4} & \frac{1}{4}
\end{pmatrix}, X_0 =
\begin{pmatrix}
1 \\ 0 \\ 0 \\ 0
\end{pmatrix}$$


##El objetivo es calcular y graficar las probabilidades de $x_n, y_n, z_n, t_n$ para varios valores de <font color = red4>$n$,
##y determinar si todos los sectores se vuelven igual de probables con el paso del tiempo.

```python

import numpy as np
import matplotlib.pyplot as plt

# ---Matriz de transición A ---
A = np.array([
    [1/4, 0,   1/4, 2/4],
    [2/4, 1/4, 0,   1/4],
    [1/4, 2/4, 1/4, 0  ],
    [0,   1/4, 2/4, 1/4]
])

# ---Vector inicial ---
# Supongamos que al inicio todo está en el primer sector
X0 = np.array([1, 0, 0, 0])

# ---Número de iteraciones ---
n = 50

# ---Listas para guardar la evolución ---
x_vals, y_vals, z_vals, t_vals = [], [], [], []

# --- Iteración: Xn = A * X_{n-1} ---
X = X0
for i in range(n):
    X = A @ X    # multiplicación matricial
    x_vals.append(X[0])
    y_vals.append(X[1])
    z_vals.append(X[2])
    t_vals.append(X[3])

# --- . Graficar resultados ---
plt.figure(figsize=(8,5))
plt.plot(x_vals, label='x_n', marker='o')
plt.plot(y_vals, label='y_n', marker='s')
plt.plot(z_vals, label='z_n', marker='^')
plt.plot(t_vals, label='t_n', marker='d')
plt.title('Evolución de las probabilidades en el tiempo')
plt.xlabel('Iteraciones (n)')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.legend()
plt.show()

# --- Imprimir valores finales ---
print("Valores finales aproximados:")
print(f"x_n ≈ {x_vals[-1]:.4f}")
print(f"y_n ≈ {y_vals[-1]:.4f}")
print(f"z_n ≈ {z_vals[-1]:.4f}")
print(f"t_n ≈ {t_vals[-1]:.4f}")
```

## <font color = red> Interpretacion de los resultados

#Al observar la gráfica, las curvas de $x_n, y_n, z_n, t_n$ convergen al mismo valor conforme aumenta <font color = red4>$n$.

##Este valor $n$ representa la distribución estacionaria del sistema, donde las probabilidades dejan de cambiar  


##En términos numéricos, los resultados finales son aproximadamente:

$$x_n \approx y_n \approx z_n \approx t_n \approx 0.25$$
