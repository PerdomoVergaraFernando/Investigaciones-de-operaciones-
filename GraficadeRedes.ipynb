#<font color = Purple> 2.- Crear un Grafo vacio.

## Tipos de Graficos

Graph(): Funcion para crea un grafo no dirigido.

DiGraph(): Funcion para crea un grafo dirigido.

MultiGraph(): Funcion para crea un grafo no dirigido con múltiples aristas entre nodos.

```python
G=nx.Graph() #Se lo asigmanos a una variable para poder trabajar mejor.
```

#<font color = Purple> 3.- Agregar Nodos.

add_node(nodo): Funcion que agrega un único nodo.

add_nodes_from(lista): Funcion que permite añadir varios nodos de golpe.

Los nodos pueden llevar atributos, como color, peso, nombre, etc. (se guardan como diccionarios asociados al nodo).

vea el siguiente ejemplo:


```python
# Agregar un nodo individual
G.add_node(1)

# Agregar varios nodos a la vez
G.add_nodes_from([2, 3])

# Agregar nodos con atributos (ejemplo: color)
G.add_nodes_from([(4, {"color": "rojo"}), (5, {"color": "verde"})])

nx.draw_networkx(G)
```

#<font color = Purple> 4.- Agregar Aristas.

Una arista conecta dos nodos.

add_edge(a, b): conecta a con b.

add_edges_from(lista): permite meter muchas aristas a la vez.

En este caso, ahora tenemos un grafo con 5 nodos y varias conexiones.

vea el siguiente ejemplo:

```python
# Agregar una arista entre 1 y 2
G.add_edge(1, 2)

# Agregar usando tupla
e = (2, 3)
G.add_edge(*e)

# Agregar múltiples aristas
G.add_edges_from([(1, 3), (3, 4), (4, 5)])
nx.draw_networkx(G)
```

#<font color = Purple>  5.- Explorar el Grafo.

G.nodes: Funcion que devuelve los nodos.

G.edges: Funcion que devuelve las conexiones.

G.adj[nodo]: Funcion que muestra los vecinos de un nodo.

G.degree[nodo]: Funcion que muestra cuántas aristas tiene ese nodo.

nx.number_of_nodes(): Está función me devuelve la cantidad de nodos.

nx.number_of_edges() # Está función me devuelve la cantidad de Aristas (Arcos, Ramas o Ligaduras).

nx.shortest_path(G, source, target, weight=None): Funcion que me da el camino mas corto del Grafo.

      G: el grafo.

      source: nodo de inicio.

      target: nodo de destino.

      weight: si quieres usar pesos (ej: distancia, tiempo, costo).

      weight=None → usa la cantidad de aristas (camino más corto en saltos).

      weight="nombre_atributo" → usa el atributo que hayas definido en las aristas (ej: "distancia").

vea el siguiente ejemplo:

```python
# Listar nodos y aristas
print("Nodos:", G.nodes)
print("Aristas:", G.edges)

# Ver nodos adyacentes a 1
print("Vecinos de 1:", list(G.adj[1]))

# Ver el grado de un nodo (cuántas conexiones tiene)
print("Grado de 1:", G.degree[1])

print("Cantidad de nodos", nx.number_of_nodes(G)) #Está función me devuelve la cantidad de nodos

print("Cantidad de Aristas",nx.number_of_edges(G)) # Está función me devuelve la cantidad de Aristas (Arcos, Ramas o Ligaduras)

print("El camino mas corto", nx.shortest_path(G, 1, 5, weight=None))
```

#<font color = Purple>  6.- Eliminar Nodos y Aristas

Puedes borrar nodos y todas sus aristas.

También puedes borrar solo ciertas conexiones sin eliminar el nodo

remove_node(nodo): Funcion que elimina un nodo.

G.remove_nodes_from([a, b]): Funcion que elimina varios nodos.

G.remove_edge(a, b): Funcion que elimina una arista.

vea el siguiente ejemplo:

```python
# Eliminar un nodo
G.remove_node(5)

# Eliminar varios nodos
# Check if nodes 2 and 3 still exist before attempting to remove them
nodes_to_remove = [node for node in [2, 3] if node in G]
G.remove_nodes_from(nodes_to_remove)

# Eliminar una arista
# Check if the edge (1, 3) still exists before attempting to remove it
if G.has_edge(1, 3):
    G.remove_edge(1, 3)

nx.draw_networkx(G)
```

#<font color = Purple>  6.- Constrir Grafos de otra forma.

En lugar de agregar nodos y aristas paso a paso, también podemos crear grafos directamente:

Graph([(a, b), (b, c), ...]) Funcion que construye un grafo directamente con las conexiones.

Graph(diccionario): Funcion que usa un diccionario de adyacencia, donde cada clave es un nodo y su valor es la lista de vecinos.

vea los siguientes ejemplos:

```python
# Grafo a partir de una lista de aristas
H = nx.Graph([(0, 1), (1, 2), (2, 3)])

nx.draw_networkx(H)
```

```python
# Grafo desde diccionario de adyacencia
adj_dict = {0: [1, 2], 1: [0, 2], 2: [0, 1]}
J = nx.Graph(adj_dict)
nx.draw_networkx(J) #Funcion que dibuja el grafo.
```

# <font color= red> Ejemplo práctico: Red de Ciudades y Rutas

Imagina que tenemos varias ciudades y queremos representar las rutas de carretera entre ellas.

```python
## 1.- Crear el grafo:

import networkx as nx
import matplotlib.pyplot as plt

# Creamos un grafo no dirigido
G = nx.Graph()

## 2.- Agregar nodos (ciudades):

# Agregar nodos con atributos (población en millones)
G.add_nodes_from([
    ("Ciudad de México", {"poblacion": 9}),
    ("Guadalajara", {"poblacion": 1.5}),
    ("Monterrey", {"poblacion": 1.1}),
    ("Puebla", {"poblacion": 1.7}),
    ("Querétaro", {"poblacion": 0.9})
])

## 3.- Agregar aristas (rutas con distancias):

# Agregar rutas con su distancia en km
G.add_edges_from([
    ("Ciudad de México", "Guadalajara", {"distancia": 550}),
    ("Ciudad de México", "Monterrey", {"distancia": 900}),
    ("Ciudad de México", "Puebla", {"distancia": 130}),
    ("Ciudad de México", "Querétaro", {"distancia": 220}),
    ("Guadalajara", "Monterrey", {"distancia": 700}),
    ("Puebla", "Querétaro", {"distancia": 280})
])

## 4.- Explorar el grafo:

# Ver nodos y sus atributos
print("Nodos:", G.nodes(data=True))

# Ver aristas con sus atributos
print("Aristas:", G.edges(data=True))

# Ver vecinos de Ciudad de México
print("Vecinos de Ciudad de México:", list(G.adj["Ciudad de México"]))

# Ver grado de Guadalajara (cuántas rutas salen de ahí)
print("Grado de Guadalajara:", G.degree["Guadalajara"])

## 5.- Eliminar nodos y aristas:

# Supongamos que cerramos la ruta Puebla-Querétaro
G.remove_edge("Puebla", "Querétaro")

# Y Querétaro queda fuera de la red
G.remove_node("Querétaro")

## 6.- Visualización del grafo:

# Dibujar el grafo
pos = nx.spring_layout(G, seed=42)  # Layout para la posición de los nodos

# Dibujar nodos y aristas
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=2000, font_size=10, font_weight="bold")

# Dibujar etiquetas con las distancias en las aristas
labels = nx.get_edge_attributes(G, "distancia")
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()

## 7.- Camino mas corto:

# Camino más corto en número de saltos
camino1 = nx.shortest_path(G, source="Ciudad de México", target="Monterrey")
print("Camino más corto en saltos:", camino1)

# Camino más corto en distancia (usando el atributo 'distancia')
camino2 = nx.shortest_path(G, source="Ciudad de México", target="Monterrey", weight="distancia")
print("Camino más corto en distancia:", camino2)

# Distancia total del camino más corto
distancia_total = nx.shortest_path_length(G, source="Ciudad de México", target="Monterrey", weight="distancia")
print("Distancia total:", distancia_total, "km")

```

# <font color= Purple> 1.- Ejemplo de Árbol de expansión mínima

```python

# Árbol de expansión mínima (usando las distancias como peso)
mst = nx.minimum_spanning_tree(G, weight="distancia")

print("Aristas del Árbol de Expansión Mínima:")
print(mst.edges(data=True))
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=2000, font_size=10, font_weight="bold")
```

# <font color= Purple> 2.- Ejemplo de Ruta más corta

```python
# Camino más corto en distancia entre Ciudad de México y Monterrey
camino = nx.shortest_path(G, source="Ciudad de México", target="Monterrey", weight="distancia")
distancia = nx.shortest_path_length(G, source="Ciudad de México", target="Monterrey", weight="distancia")

print("Ruta más corta:", camino)
print("Distancia total:", distancia, "km")
```

# <font color= Purple> 3.- Ejemplo de Flujo máximo

```python
# Crear un grafo dirigido con capacidades
FG = nx.DiGraph()

FG.add_edge("Ciudad de México", "Guadalajara", capacity=100)
FG.add_edge("Ciudad de México", "Monterrey", capacity=80)
FG.add_edge("Ciudad de México", "Puebla", capacity=50)
FG.add_edge("Puebla", "Monterrey", capacity=40)
FG.add_edge("Guadalajara", "Monterrey", capacity=70)

# Calcular flujo máximo de Ciudad de México a Monterrey
flow_value, flow_dict = nx.maximum_flow(FG, "Ciudad de México", "Monterrey")

print("Flujo máximo de Ciudad de México a Monterrey:", flow_value)
print("Detalle del flujo:", flow_dict)
```
