```python
import matplotlib.pyplot as plt
import numpy as np

p=1
q=0
n=25

p_i=[]
q_i=[]
for i in range(n):
  pi = 0.6*p + 0.3*q
  qi = 0.4*p + 0.7*q
  p_i.append(pi)
  q_i.append(qi)
  p=pi
  q=qi
```

```python
plt.plot(p_i)
plt.plot(q_i)
plt.show()
```
